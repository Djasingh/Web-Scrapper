```python
from googlesearch import search 
from bs4 import BeautifulSoup as bsoup
import requests
import sys
```

```python
def google_scrape(url):
    title=''
    text=''
    try:
        page = requests.get(url)
        if page.status_code==200:
            soup = bsoup(page.content, 'html.parser')
            try:
                title += soup.find('title').getText()
            except AttributeError as error:
                print(error)
            for script in soup(["script", "style"]):
                script.extract()    # rip it out
                print('extract')
            # get text
            text += soup.get_text()
            # break into lines and remove leading and trailing space on each
            lines = (line.strip() for line in text.splitlines())
            # break multi-headlines into a line each
            chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
            # drop blank lines
            text = '\n'.join(chunk for chunk in chunks if chunk)
            
        else:
            print("HTTP Status " + str(page.status_code) + "\n")       
    except requests.exceptions.SSLError:
        print("Oops!",sys.exc_info()[0],"occured.")
        print("Next entry.")
        print()
    except OSError as e:
        print(e)
        print()
    return title, text, page
```

```python
query = 'Alan turing AND Alonzo Church'
for url in search(query, stop=10):
    if 'YouTube' not in url:
        print(url)
        a,b= google_scrape(url)
        print('Title: {}'.format(a))
        print('Text:{}'.format(b))
        if 'Publications' in b:
            print("==================================================")
            print('Publications:{}'.format(b.split('Publications')[1]))
            print("==================================================")
        print("---------------------------------------------------------")
```
